# **Binary Search / Greedy**
***MinMaxDivision***

The task:

Split the array into K blocks (consecutive subarrays).

Each block can be empty, but all elements of A must be assigned to some block.

The large sum is the maximum block sum.

Goal: minimize this large sum.

This is solved using binary search + greedy check.

```python
function solution(K, M, A):
    low  = max(A)             # minimal possible large sum (at least the max element)
    high = sum(A)             # maximal possible large sum (everything in one block)
    result = high

    while low <= high:
        mid = (low + high) // 2
        if canDivide(A, K, mid):
            result = mid
            high = mid - 1    # try smaller max sum
        else:
            low = mid + 1     # need bigger max sum

    return result


function canDivide(A, K, maxBlockSum):
    countBlocks = 1
    currentSum = 0
    for each value in A:
        if currentSum + value > maxBlockSum:
            countBlocks += 1
            currentSum = value
            if countBlocks > K:
                return False
        else:
            currentSum += value
    return True
```
